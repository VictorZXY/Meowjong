```python
import os
import pandas as pd
from matplotlib import pyplot as plt

RESULTS_DIR = '..\\..\\results\\REINFORCE-evaluation\\'
CSV_FILENAME = 'REINFORCE-evaluation.csv'
```

```python
def load_data(path, file):
    csv_path = os.path.join(path, file)
    return pd.read_csv(csv_path, sep=',')
```

```python
rl_eval = load_data(RESULTS_DIR, CSV_FILENAME)
rl_eval
```

```python
max(rl_eval['total_win_rate'])
```

```python
plt.plot(rl_eval['eps'], rl_eval['east_win_rate'])
plt.plot(rl_eval['eps'], rl_eval['south_win_rate'])
plt.plot(rl_eval['eps'], rl_eval['west_win_rate'])
plt.plot(rl_eval['eps'], rl_eval['total_win_rate'])

plt.show()
```

```python
plt.plot(rl_eval['eps'], rl_eval['east_avg_score'] * (3.5 / 3) / 1.5)
plt.plot(rl_eval['eps'], rl_eval['south_avg_score'] * (3.5 / 3))
plt.plot(rl_eval['eps'], rl_eval['west_avg_score'] * (3.5 / 3))
plt.plot(rl_eval['eps'], rl_eval['total_avg_score'])

plt.show()
```
